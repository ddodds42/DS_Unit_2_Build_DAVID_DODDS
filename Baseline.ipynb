## Baseline predictions of Human Development and State Failure events are meant to emulate a human making an educated guess simply based on a nation-states performance in the most recent years, according to this dataset.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
df = pd.read_excel('Master_MELT.xlsx')
df
```

```python
df.describe()
```

```python
df.head(1)
```

## The dataframe created below aligns last year's state failure events with state failures predicted 5 years from now.

```python
State_fail_yr_5 = df[df.year0 < 2014].iloc[:,4:5]
State_fail_yr_5.columns=['predicted']
ninety_19 = df[df.year0 > 1989]
ninety_14 = ninety_19[ninety_19['year0'] < 2015]
State_fail_yr_5['actual'] = np.array(ninety_14.iloc[:,6:7])
State_fail_yr_5
```

```python
f1_score(State_fail_yr_5['actual'], State_fail_yr_5['predicted'])
```

## BASELINE: T_State_Fail_yr_1 (state failure on year one), predicted by state failure in year 0, an F-Score of 0.80547.

## BASELINE: T_State_Fail_yr_5 (state failure on year five), predicted by state failure in year 0, an F-Score of 0.49036.

```python
State_fail_nxt_5 = df[df.year0 < 2014].iloc[:,4:5]
State_fail_nxt_5.columns=['predicted']
ninety_19 = df[df.year0 > 1989]
ninety_14 = ninety_19[ninety_19['year0'] < 2015]
State_fail_nxt_5['actual'] = np.array(ninety_14.iloc[:,5:6])
State_fail_nxt_5
```

## The dataframe above aligns last year's state failure events with state failures predicted in any of the next 5 years.

```python
f1_score(State_fail_nxt_5['actual'], State_fail_nxt_5['predicted'])
```

## ^^ BASELINE: T_State_Fail_plus_5 (state failure in any of the next 5 years), predicted by state failure in year 0, an F-Score of 0.6685 ^^

```python
HDI_Mean = df[df.year0 < 2012].iloc[:,2:3]
HDI_Mean.columns=['predicted']

eighty9_16 = df[df.year0 < 2017]
ninety4_16 = eighty9_16[eighty9_16.year0 > 1993]

HDI_Mean['actual'] = np.array(ninety4_16.iloc[:,2:3])
HDI_Mean
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
(mean_squared_error(HDI_Mean['actual'], HDI_Mean['predicted']))**0.5
```

```python
r2_score(HDI_Mean['actual'], HDI_Mean['predicted'])
```
