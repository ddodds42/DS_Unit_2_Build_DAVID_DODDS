```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel('Master_MELT.xlsx')
df
```

## Dataframe "Present" represents all years which currently have complete 5 year HDI and State Failure Event forcast data: 2014 and before.

```python
present = df[df.year0 < 2015]
present
```

## Dataframe "Forecast" represents all years which do not have complete 5 year HDI and State Failure Event forcast data: 2015 and after.

```python
forecast = df[df.year0 > 2014]
forecast
```

## Dataframe "semi_forecast" represents all years for which we have partial 5-year HDI and State Failure Event forcast data: 2015 - 2018. It can be used to evaluate the model as the early 2020's unfurl; we do not have to wait until 2025 to evaluate the model's validity. 2015 data can be used to evaluate the performance model in late 2020, 2016 data can be used to evaluate the performance model in late 2021, so on and so forth.

```python
semi_forecast = forecast[forecast.year0 < 2019]
semi_forecast
```

## Dataframe "forecast_20s" represents 2019 data, which, at the time of data-gathering, has no HDI and State Failure Event forcast data. It is the test set that will not be completely evaluatable until 2025. Outputs from this dataframe are to predict nation state prosperity or failure events in the early 2020's.

```python
forecast_20s = forecast[forecast.year0 > 2018]
forecast_20s
```

## The "explore" dataset is the source of the training and the validation data. It is to be used in its entirety for cross validation models. Explore and test data are stratified by year, to guarantee that a roughly equal number of each year's data occur in each set after random splitting.

```python
explore, test = train_test_split(present, test_size=0.15, random_state=42, stratify=present['year0'])
explore.shape, test.shape
```

```python
train, val = train_test_split(explore, test_size=695, random_state=42, stratify=explore['year0'])
train.shape, val.shape
```
